```python
import torch
import torch.utils.data as Data
import torch.nn.init as init
import torch.optim as optim
import numpy as np
import sklearn.preprocessing as preprocessing
```

### 制作数据集

```python
t1 = np.array([0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0])
t2 = np.array([35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4])
scaler = preprocessing.StandardScaler()
#scaler2 = preprocessing.StandardScaler().fit(t2.reshape(-1,1))
#t1 = scaler1.transform(t1.reshape(-1,1))
#t2 = scaler2.transform(t2.reshape(-1,1))
t1 = scaler.fit_transform(t1.reshape(-1, 1))
t2 = scaler.fit_transform(t2.reshape(-1,1))
batch_size = t1.shape[0]
print(batch_size)
dataset = Data.TensorDataset(torch.tensor(t1).float(), torch.tensor(t2).float())
dataset = Data.DataLoader(dataset, batch_size=batch_size)
for x, y in dataset:
    print(x, y)
```

### 定义模型

```python
weight_nums = 1
net = torch.nn.Sequential()
net.add_module('linear', torch.nn.Linear(weight_nums, 1))  # torch.nn.Linear(in_features, out_features, bias=True)
print(net)
for param in net.parameters():
    print(param)
```

### 初始化模型参数

```python
init.normal_(net[0].weight, mean=0, std=0.1)
init.constant_(net[0].bias, val=0)
for param in net.parameters():
    print(param)
```

### 定义损失函数

```python
loss = torch.nn.MSELoss()
```

### 定义优化算法

```python
optimizer = optim.SGD(net.parameters(), lr=0.0001)
print(optimizer)
```

### 训练模型

```python
def training_loop(n_epochs, dataset, network, loss, optimizer):
    for epoch in range(n_epochs + 1):
        for x, y in dataset:
            output = network(x)
            l = loss(output, y.view(-1, 1))
            optimizer.zero_grad()
            l.backward()
            optimizer.step()
        print('epoch %d, loss %f' % (epoch, l.item()))  
    print(net[0].weight, net[0].bias)
```

```python
if __name__ == "__main__":
    n_epochs = 50000
    training_loop(n_epochs=n_epochs, dataset=dataset, network=net, loss=loss, optimizer=optimizer)
```
